## Problema

### - Vamos supor que precisamos encontrar a soma de inteiros de 1 a 10, de 20 a 37, e de 35 a 49.

### - Se for criado um programa para adicionar esses três conjuntos de números, o **código** pode parecer assim:

```python
soma = 0
for i in range(1, 11):
    soma += i
print("A soma de 1 a 10  é", soma)

soma = 0
for i in range(20, 38):
    soma += i
print("A soma de 20 a 37 é", soma)

soma = 0
for i in range(35, 50):
    soma += i
print("A soma de 35 a 49 é", soma)
```

## Por que funções?

### - O código para calcular essas somas é muito semelhante!

### - Não seria interessante poder escrever um código amplamente utilizado apenas uma vez e reutilizá-lo?

### - É possivel fazer isso definindo uma função, que permite criar código reutilizável?

### - Por exemplo, o código anterior pode ser simplificado usando funções, da seguinte forma:

```python
def soma(l1, l2):
    result = 0
    for i in range(l1, l2 + 1):
        result += i
    return result

def imprimir():
  print("A soma de 1 a 10 é", soma(1, 10))
  print("A soma de 20 a 37 é", soma(20, 37))
  print("A soma de 35 a 49 é", soma(35, 49))

def main():
  imprimir()

main()
```

# 2. **Funções em Python**


### - São blocos de código reutilizáveis que realizam uma tarefa específica.

### - Elas permitem que os programadores organizem o código de maneira mais modular, o que torna mais fácil entender, manter e reutilizar partes do programa.

### - Uma função pode ser chamada quantas vezes forem necessárias em diferentes partes de um programa, e pode ou não receber dados como entrada e retornar dados como saída.

### - A palavra-chave para definir funções em Python é *def*.

### - Usa-se a palavra-chave *def* da seguinte maneira:

```python
def nome_da_funcao(parâmetros_opcionais):
    # Bloco de código da função
    return valor_opcional
```

## O formato básico para definir uma função em Python:

### 1. def: Inicia a definição da função;
### 2. nome_da_funcao: Nome da função, que você escolhe;
### 3. parâmetros_opcionais: Os valores que você pode passar para a função;
### 4. Bloco de código da função: O corpo da função, que será executado quando for chamada a função;
### 5. return: (Opcional) Usado para retornar um valor da função.

## **Exemplo 06** - Pesquisa em uma lista

```python
lista = [10, 20, 25, 30]

def pesquisar(lista, valor):
    for index,value in enumerate(lista):
        if value == valor:
            return index
    return None
```

```python
print(pesquisar(lista, 30))
```

## 2.7 **Funções como objetos de primeira classe**

## Em Python, funções são objetos de primeira classe, o que significa que podem ser passadas como argumentos para outras funções, armazenadas em variáveis, ou mesmo retornadas por outras funções.

### Exemplo

## 2.8 **Funções recursivas**

### Funções recursivas são aquelas que se chamam a si mesmas.
### Elas são úteis para resolver problemas que podem ser decompostos em subproblemas menores, como cálculos de fatorial ou sequência de Fibonacci.



### Exemplo

```python
def fatorial(n):
    if n == 0:
        return 1
    else:
        return n * fatorial(n - 1)

print(fatorial(5))
```

### Neste exemplo, a função fatorial se chama repetidamente até que n seja igual a 0.

## 3 **Modularização em Python**

### A modularização em Python é a prática de dividir um programa extenso em partes menores, chamadas módulos, que são essencialmente arquivos Python separados com funções, classes ou variáveis que podem ser reutilizadas.

### Modularizar o código torna os programas mais organizados, fáceis de entender, de manter e de escalar.

### Em vez de ter um único arquivo gigantesco contendo todo o código do projeto, podemos dividir o programa em diferentes módulos, cada um com uma responsabilidade específica.

### Vantagens da modularização:

1. Organização: O código fica mais limpo e organizado, com cada módulo lidando com uma parte específica do programa.

2. Reutilização: Módulos podem ser reutilizados em diferentes projetos, evitando a necessidade de reescrever código.

3. Facilidade de manutenção: Localizar e corrigir erros é mais simples em um código modularizado, já que o problema estará restrito a um único módulo.

4. Colaboração: Em projetos grandes, diferentes desenvolvedores podem trabalhar simultaneamente em módulos diferentes.

## Estrutura de um projeto modularizado - Soma

### Suponha que estamos desenvolvendo um sistema de gerenciamento de tarefas.

###Podemos dividir o projeto em diferentes módulos como mostrado abaixo:

```python
meu_projeto/
│
├── principal.py
├── entrada.py
```

### Projeto Soma

```python
#entrada.py
def valida_inteiro(mensagem, minimo, maximo):
	while True:
		try:
			v = int(input(mensagem))
			if v >= minimo and v <= maximo:
				return v
			else:
				print("Digite um valor entre %d e %d" % (minimo, maximo))
		except:
				print("Vecê deve digitar um número inteiro")
```

```python
#principal.py
from entrada import valida_inteiro
lista = []
for x in range(5):
	lista.append(valida_inteiro("Digite um número:", 0, 20))
print("Soma: %d" % (sum(lista)))
```

## Estrutura de um projeto modularizado Tarefas

### Suponha que estamos desenvolvendo um sistema de gerenciamento de tarefas.

###Podemos dividir o projeto em diferentes módulos como mostrado abaixo:

```python
meu_projeto/
│
├── principal.py
├── modulo_usuarios.py
└── modulo_tarefas.py
```

* principal.py: Arquivo principal que executa o programa.

* modulo_usuarios.py: Lida com todas as operações relacionadas aos usuários (criação, autenticação, etc.).

* modulo_tarefas.py: Lida com operações relacionadas às tarefas (criação, edição, exclusão, etc.).

## **Exemplo de implementação modularizada**

1. modulo_usuarios.py: Operações relacionadas a usuários

```python
# modulo_usuarios.py

def criar_usuario(nome, senha):
    return {"nome": nome, "senha": senha}

def autenticar_usuario(usuario, senha):
    if usuario["senha"] == senha:
        return True
    else:
        return False
```

2. modulo_tarefas.py: Operações relacionadas a tarefas

```python
# modulo_tarefas.py

def adicionar_tarefa(tarefas, descricao):
    tarefas.append(descricao)
    return tarefas

def listar_tarefas(tarefas):
    for idx, tarefa in enumerate(tarefas, 1):
        print(f"{idx}. {tarefa}")
```

3. principal.py: Arquivo principal que coordena o programa

```python
# principal.py

from modulo_usuarios import criar_usuario, autenticar_usuario
from modulo_tarefas import adicionar_tarefa, listar_tarefas

def main():
    # Criando um usuário
    usuario = criar_usuario("admin", "1234")

    # Autenticando usuário
    senha = input("Digite sua senha: ")

    if autenticar_usuario(usuario, senha):
        print("Autenticação bem-sucedida!")

        # Gerenciando tarefas
        tarefas = []
        while True:
            print("\n1. Adicionar tarefa")
            print("2. Listar tarefas")
            print("3. Sair")
            opcao = input("Escolha uma opção: ")

            if opcao == "1":
                descricao = input("Digite a descrição da tarefa: ")
                adicionar_tarefa(tarefas, descricao)
            elif opcao == "2":
                listar_tarefas(tarefas)
            elif opcao == "3":
                print("Saindo...")
                break
            else:
                print("Opção inválida!")

    else:
        print("Autenticação falhou!")

if __name__ == "__main__":
    main()
```
