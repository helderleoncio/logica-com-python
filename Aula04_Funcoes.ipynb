## Problema

### - Vamos supor que precisamos encontrar a soma de inteiros de 1 a 10, de 20 a 37, e de 35 a 49.

### - Se for criado um programa para adicionar esses trs conjuntos de nmeros, o **cdigo** pode parecer assim:

```python
soma = 0
for i in range(1, 11):
    soma += i
print("A soma de 1 a 10  ", soma)

soma = 0
for i in range(20, 38):
    soma += i
print("A soma de 20 a 37 ", soma)

soma = 0
for i in range(35, 50):
    soma += i
print("A soma de 35 a 49 ", soma)
```

## Por que funes?

### - O cdigo para calcular essas somas  muito semelhante!

### - No seria interessante poder escrever um cdigo amplamente utilizado apenas uma vez e reutiliz-lo?

### -  possivel fazer isso definindo uma funo, que permite criar cdigo reutilizvel?

### - Por exemplo, o cdigo anterior pode ser simplificado usando funes, da seguinte forma:

```python
def soma(l1, l2):
    result = 0
    for i in range(l1, l2 + 1):
        result += i
    return result

def imprimir():
  print("A soma de 1 a 10 ", soma(1, 10))
  print("A soma de 20 a 37 ", soma(20, 37))
  print("A soma de 35 a 49 ", soma(35, 49))

def main():
  imprimir()

main()
```

# 2. **Funes em Python**

## 2.8 **Funes recursivas**

### Funes recursivas so aquelas que se chamam a si mesmas.
### Elas so teis para resolver problemas que podem ser decompostos em subproblemas menores, como clculos de fatorial ou sequncia de Fibonacci.



### Exemplo

```python
def fatorial(n):
    if n == 0:
        return 1
    else:
        return n * fatorial(n - 1)

print(fatorial(5))
```

### Neste exemplo, a funo fatorial se chama repetidamente at que n seja igual a 0.

## 3 **Modularizao em Python**

### A modularizao em Python  a prtica de dividir um programa extenso em partes menores, chamadas mdulos, que so essencialmente arquivos Python separados com funes, classes ou variveis que podem ser reutilizadas.

### Modularizar o cdigo torna os programas mais organizados, fceis de entender, de manter e de escalar.

### Em vez de ter um nico arquivo gigantesco contendo todo o cdigo do projeto, podemos dividir o programa em diferentes mdulos, cada um com uma responsabilidade especfica.

### Vantagens da modularizao:

1. Organizao: O cdigo fica mais limpo e organizado, com cada mdulo lidando com uma parte especfica do programa.

2. Reutilizao: Mdulos podem ser reutilizados em diferentes projetos, evitando a necessidade de reescrever cdigo.

3. Facilidade de manuteno: Localizar e corrigir erros  mais simples em um cdigo modularizado, j que o problema estar restrito a um nico mdulo.

4. Colaborao: Em projetos grandes, diferentes desenvolvedores podem trabalhar simultaneamente em mdulos diferentes.

## Estrutura de um projeto modularizado - Soma

```python
#entrada.py
def valida_inteiro(mensagem, minimo, maximo):
	while True:
		try:
			v = int(input(mensagem))
			if v >= minimo and v <= maximo:
				return v
			else:
				print("Digite um valor entre %d e %d" % (minimo, maximo))
		except:
				print("Vec deve digitar um nmero inteiro")
```

## Estrutura de um projeto modularizado Tarefas

### Suponha que estamos desenvolvendo um sistema de gerenciamento de tarefas.

###Podemos dividir o projeto em diferentes mdulos como mostrado abaixo:

```python
meu_projeto/

 principal.py
 modulo_usuarios.py
 modulo_tarefas.py
```

* principal.py: Arquivo principal que executa o programa.

* modulo_usuarios.py: Lida com todas as operaes relacionadas aos usurios (criao, autenticao, etc.).

* modulo_tarefas.py: Lida com operaes relacionadas s tarefas (criao, edio, excluso, etc.).

## **Exemplo de implementao modularizada**

1. modulo_usuarios.py: Operaes relacionadas a usurios

```python
# modulo_usuarios.py

def criar_usuario(nome, senha):
    return {"nome": nome, "senha": senha}

def autenticar_usuario(usuario, senha):
    if usuario["senha"] == senha:
        return True
    else:
        return False
```

2. modulo_tarefas.py: Operaes relacionadas a tarefas

```python
# modulo_tarefas.py

def adicionar_tarefa(tarefas, descricao):
    tarefas.append(descricao)
    return tarefas

def listar_tarefas(tarefas):
    for idx, tarefa in enumerate(tarefas, 1):
        print(f"{idx}. {tarefa}")
```

3. principal.py: Arquivo principal que coordena o programa

```python
# principal.py

from modulo_usuarios import criar_usuario, autenticar_usuario
from modulo_tarefas import adicionar_tarefa, listar_tarefas

def main():
    # Criando um usurio
    usuario = criar_usuario("admin", "1234")

    # Autenticando usurio
    senha = input("Digite sua senha: ")

    if autenticar_usuario(usuario, senha):
        print("Autenticao bem-sucedida!")

        # Gerenciando tarefas
        tarefas = []
        while True:
            print("\n1. Adicionar tarefa")
            print("2. Listar tarefas")
            print("3. Sair")
            opcao = input("Escolha uma opo: ")

            if opcao == "1":
                descricao = input("Digite a descrio da tarefa: ")
                adicionar_tarefa(tarefas, descricao)
            elif opcao == "2":
                listar_tarefas(tarefas)
            elif opcao == "3":
                print("Saindo...")
                break
            else:
                print("Opo invlida!")

    else:
        print("Autenticao falhou!")

if __name__ == "__main__":
    main()
```
