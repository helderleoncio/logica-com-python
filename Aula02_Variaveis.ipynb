### Voc investiu R\$ 1000,00 em uma aplicao financeira que cresce 5% ao ano. Qual ser o valor desse investimento aps 3 anos, sabendo que ele segue a frmula de juros compostos?

Frmula:

=(1+)A=P(1+r) t

*   o valor futuro
*   o valor inicial

## 2.2 **Problema1: Gesto de Alunos e Notas em uma Turma**

### Imagine que voc  um professor e tem uma turma de 5 alunos. Voc deseja criar um programa em Python que armazene os nomes dos alunos e suas respectivas notas em uma prova. Depois, voc precisa calcular a mdia da turma, encontrar a maior nota e o aluno que a obteve.

### **Algoritmo**
#### 1. Armazene os nomes dos alunos em uma lista.
#### 2. Armazene as notas desses alunos em outra lista.
#### 3. Calcule a mdia das notas.
#### 4. Encontre a maior nota e o aluno que a obteve.

```python
# Listas com os nomes dos alunos e suas respectivas notas
alunos = ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"]
notas = [9.5, 9.5, 9.5, 9.5, 9.5]
```

```python
# Passo 3: Calcular a mdia da turma
media = sum(notas) / len(notas)
print(f"A mdia da turma : {media:.2f}")
```

```python
# Passo 4: Encontrar a maior nota e o aluno que a obteve
maior_nota = max(notas)
indice_maior_nota = notas.index(maior_nota)
aluno_maior_nota = alunos[indice_maior_nota]
```

```python
print(f"A maior nota foi {maior_nota}, obtida pelo aluno(a): {aluno_maior_nota}")
```

## 2.2 **Tuplas**

### Uma tupla em Python  uma estrutura de dados semelhante a uma lista, mas com uma diferena fundamental: tuplas so imutveis.

### Isso significa que, uma vez criadas, os valores de uma tupla no podem ser modificados (adicionar, remover ou alterar elementos).

### Tuplas so frequentemente usadas para armazenar dados que no devem ser alterados durante a execuo do programa.

### Caractersticas das Tuplas:

#### Imutabilidade: Uma vez criada, os elementos da tupla no podem ser alterados.
### Aceitam mltiplos tipos de dados: Assim como as listas, as tuplas podem conter diferentes tipos de dados.
### Indexao: Os elementos de uma tupla podem ser acessados usando ndices, da mesma forma que em uma lista.
### Mais rpidas que listas: Como tuplas so imutveis, elas so mais eficientes em termos de desempenho.

## Definio de uma Tupla:

###Uma tupla  definida usando parnteses () e seus elementos so separados por vrgulas

```python
minha_tupla = (10, 20, 30)
```

## Acessando Elementos:
## Podemos acessar os elementos de uma tupla usando seus ndices, da mesma forma que faria com uma lista:

```python
print(minha_tupla[0])  # Sada: 10
print(minha_tupla[2])  # Sada: 30
```

## Principais Caractersticas:

### Tuplas podem conter diferentes tipos de dados:

```python
tupla_mista = (1, "texto", 3.14, True)
print(tupla_mista)  # Sada: (1, "texto", 3.14, True)
```

### Imutabilidade: No  possvel alterar os elementos de uma tupla. Tentar fazer isso resulta em erro:

```python
minha_tupla[1] = 40  # Isso resultar em erro!
```

### Fatiamento (Slicing): Voc pode acessar uma parte da tupla usando fatiamento (como em listas):

```python
print(minha_tupla[1:3])  # Sada: (20, 30)
```

### Tuplas com um nico elemento: Para criar uma tupla com um nico elemento,  necessrio adicionar uma vrgula aps o elemento, seno Python interpretar como um valor simples:

```python
tupla_com_um_elemento = (10,)
print(type(tupla_com_um_elemento))  # Sada: <class 'tuple'>
```

### Mtodos Comuns para Tuplas:
### Embora sejam imutveis, algumas operaes e funes podem ser usadas com tuplas:

#### count(): Retorna o nmero de vezes que um elemento aparece na tupla.

```python
tupla = (1, 2, 2, 3)
print(tupla.count(2))  # Sada: 2
```

#### index(): Retorna o ndice do primeiro elemento encontrado.

```python
print(tupla.index(3))  # Sada: 3
```

#### len(): Retorna o nmero de elementos na tupla.

```python
print(len(tupla))  # Sada: 4
```

#### Desempacotamento de Tuplas:  possvel atribuir os elementos de uma tupla diretamente a variveis.

```python
tupla = (10, 20, 30)
a, b, c = tupla
print(a)  # Sada: 10
print(b)  # Sada: 20
print(c)  # Sada: 30
```

## Exemplo Completo

```python
# Definindo uma tupla
tupla = ("ma", "banana", "laranja")

# Acessando elementos
print(tupla[0])  # Sada: ma

# Tentando modificar um elemento (vai resultar em erro)
# tupla[1] = "uva"  # Erro: 'tuple' object does not support item assignment

# Usando mtodos
print(tupla.count("banana"))  # Sada: 1
print(tupla.index("laranja"))  # Sada: 2

# Desempacotamento
fruta1, fruta2, fruta3 = tupla
print(fruta1)  # Sada: ma
```

### Tuplas so teis quando voc precisa armazenar conjuntos de dados constantes e imutveis no seu programa.
