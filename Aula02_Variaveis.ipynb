## 3.1.4.2.3 **Problema 03**

### Você investiu R\$ 1000,00 em uma aplicação financeira que cresce 5% ao ano. Qual será o valor desse investimento após 3 anos, sabendo que ele segue a fórmula de juros compostos?

Fórmula:

𝐴=𝑃×(1+𝑟)𝑡A=P×(1+r) t

Onde:

* 𝐴 é o valor futuro
* 𝑃 é o valor inicial
* 𝑟 é a taxa de juros (5% = 0.05)
* 𝑡 é o tempo (em anos)

```python
# Valores iniciais
valor_inicial = 1000
taxa_juros = 0.05
anos = 3
```

```python
# Cálculo do valor futuro usando potenciação
valor_futuro = valor_inicial * (1 + taxa_juros) ** anos
```

```python
# Exibindo o valor futuro
print(f"O valor do investimento após {anos} anos será: R$ {valor_futuro:.2f}")
```

## 2.2 **Problema1: Gestão de Alunos e Notas em uma Turma**

### Imagine que você é um professor e tem uma turma de 5 alunos. Você deseja criar um programa em Python que armazene os nomes dos alunos e suas respectivas notas em uma prova. Depois, você precisa calcular a média da turma, encontrar a maior nota e o aluno que a obteve.

### **Algoritmo**
#### 1. Armazene os nomes dos alunos em uma lista.
#### 2. Armazene as notas desses alunos em outra lista.
#### 3. Calcule a média das notas.
#### 4. Encontre a maior nota e o aluno que a obteve.

```python
# Listas com os nomes dos alunos e suas respectivas notas
alunos = ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"]
notas = [9.5, 9.5, 9.5, 9.5, 9.5]
```

```python
# Passo 3: Calcular a média da turma
media = sum(notas) / len(notas)
print(f"A média da turma é: {media:.2f}")
```

```python
# Passo 4: Encontrar a maior nota e o aluno que a obteve
maior_nota = max(notas)
indice_maior_nota = notas.index(maior_nota)
aluno_maior_nota = alunos[indice_maior_nota]
```

```python
print(f"A maior nota foi {maior_nota}, obtida pelo aluno(a): {aluno_maior_nota}")
```

## 2.2 **Tuplas**

### Uma tupla em Python é uma estrutura de dados semelhante a uma lista, mas com uma diferença fundamental: tuplas são imutáveis.

### Isso significa que, uma vez criadas, os valores de uma tupla não podem ser modificados (adicionar, remover ou alterar elementos).

### Tuplas são frequentemente usadas para armazenar dados que não devem ser alterados durante a execução do programa.

### Características das Tuplas:

#### Imutabilidade: Uma vez criada, os elementos da tupla não podem ser alterados.
### Aceitam múltiplos tipos de dados: Assim como as listas, as tuplas podem conter diferentes tipos de dados.
### Indexação: Os elementos de uma tupla podem ser acessados usando índices, da mesma forma que em uma lista.
### Mais rápidas que listas: Como tuplas são imutáveis, elas são mais eficientes em termos de desempenho.

## Definição de uma Tupla:

###Uma tupla é definida usando parênteses () e seus elementos são separados por vírgulas

```python
minha_tupla = (10, 20, 30)
```

## Acessando Elementos:
## Podemos acessar os elementos de uma tupla usando seus índices, da mesma forma que faria com uma lista:

```python
print(minha_tupla[0])  # Saída: 10
print(minha_tupla[2])  # Saída: 30
```

## Principais Características:

### Tuplas podem conter diferentes tipos de dados:

```python
tupla_mista = (1, "texto", 3.14, True)
print(tupla_mista)  # Saída: (1, "texto", 3.14, True)
```

### Imutabilidade: Não é possível alterar os elementos de uma tupla. Tentar fazer isso resulta em erro:

```python
minha_tupla[1] = 40  # Isso resultará em erro!
```

### Fatiamento (Slicing): Você pode acessar uma parte da tupla usando fatiamento (como em listas):

```python
print(minha_tupla[1:3])  # Saída: (20, 30)
```

### Tuplas com um único elemento: Para criar uma tupla com um único elemento, é necessário adicionar uma vírgula após o elemento, senão Python interpretará como um valor simples:

```python
tupla_com_um_elemento = (10,)
print(type(tupla_com_um_elemento))  # Saída: <class 'tuple'>
```

### Métodos Comuns para Tuplas:
### Embora sejam imutáveis, algumas operações e funções podem ser usadas com tuplas:

#### count(): Retorna o número de vezes que um elemento aparece na tupla.

```python
tupla = (1, 2, 2, 3)
print(tupla.count(2))  # Saída: 2
```

#### index(): Retorna o índice do primeiro elemento encontrado.

```python
print(tupla.index(3))  # Saída: 3
```

#### len(): Retorna o número de elementos na tupla.

```python
print(len(tupla))  # Saída: 4
```

#### Desempacotamento de Tuplas: É possível atribuir os elementos de uma tupla diretamente a variáveis.

```python
tupla = (10, 20, 30)
a, b, c = tupla
print(a)  # Saída: 10
print(b)  # Saída: 20
print(c)  # Saída: 30
```

## Exemplo Completo

```python
# Definindo uma tupla
tupla = ("maçã", "banana", "laranja")

# Acessando elementos
print(tupla[0])  # Saída: maçã

# Tentando modificar um elemento (vai resultar em erro)
# tupla[1] = "uva"  # Erro: 'tuple' object does not support item assignment

# Usando métodos
print(tupla.count("banana"))  # Saída: 1
print(tupla.index("laranja"))  # Saída: 2

# Desempacotamento
fruta1, fruta2, fruta3 = tupla
print(fruta1)  # Saída: maçã
```

## Quando Usar Tuplas?

### Quando você tem dados que não devem ser alterados ao longo do programa.
### Para aumentar a performance, pois tuplas são mais rápidas que listas em alguns contextos.
### Quando você quer garantir a integridade dos dados (por exemplo, coordenadas geográficas, registros fixos, etc.).
### Tuplas são úteis quando você precisa armazenar conjuntos de dados constantes e imutáveis no seu programa.
