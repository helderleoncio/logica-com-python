## **Estrutura if**

* A estrutura if avalia uma condição, e se essa condição for verdadeira (True), o bloco de código indentado a seguir será executado.
* Caso contrário, o Python pula esse bloco e continua a execução do programa.

```python
if condição:
    # Código a ser executado se a condição for verdadeira
```

### **Exemplo 01**

```python
idade = 18
if idade >= 18:
    print("Você é maior de idade.")
```

### **Exemplo 02**

```python
salário = float(input("Digite o salário para cálculo do imposto: "))

base = salário
imposto = 0

if base > 3000:
    imposto = imposto + ((base - 3000) * 0.35)
    base = 3000

if base > 1000:
    imposto = imposto + ((base - 1000) * 0.20)

print("Salário: R$%6.2f\nImposto a pagar: R$%6.2f" % (salário, imposto))
```

## ***Estrutura if-else***
* A estrutura if-else expande o conceito do if. Se a condição for verdadeira, um bloco é executado.
* Se for falsa, o bloco else será executado.

### **Sintaxe**

```python
if condição:
    # Código a ser executado se a condição for verdadeira
else:
    # Código a ser executado se a condição for falsa
```

### **Exemplo**

```python
idade = 16
if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

### **Exemplo 02**

```python
minutos = int(input("Quantos minutos você utilizou este mês: "))

if minutos < 200:
    preço = 0.20
else:
    if minutos < 400:
        preço = 0.18
    else:
        preço = 0.15

print("Você vai pagar este mês: R$%6.2f" % (minutos * preço))
```

## **Estrutura if-elif-else**

* Se houver múltiplas condições a serem verificadas, o elif pode ser utilizado para testar condições adicionais.
* O elif é abreviação de "else if".
* Esta estrutura permite que diferentes blocos de código sejam executados com base em várias condições.

### **Sintaxe**

```python
if condição1:
    # Código a ser executado se condição1 for verdadeira
elif condição2:
    # Código a ser executado se condição2 for verdadeira
else:
    # Código a ser executado se nenhuma condição for verdadeira
```

### **Exemplo 01**

```python
nota = 75
if nota >= 90:
    print("Você tirou um A.")
elif nota >= 80:
    print("Você tirou um B.")
elif nota >= 70:
    print("Você tirou um C.")
else:
    print("Você precisa estudar mais.")
```

# **Operadores de comparação e lógicos**

* Para criar essas condições, você pode usar operadores de comparação e lógicos. Alguns exemplos de operadores de comparação são:

* == (igual a)
* != (diferente de)
* \> (maior que)

## Operadores lógicos booleanos

São ferramentas essenciais em programação, usadas para combinar ou modificar expressões lógicas (verdadeiro ou falso). Em Python, eles são empregados para avaliar expressões e retornar um valor booleano: True (verdadeiro) ou False (falso). Esses operadores são muito úteis em estruturas de decisão, laços e outros controles de fluxo.

### Principais operadores lógicos booleanos em Python:###
1- and (E lógico): O operador and retorna True apenas se ambas as condições forem verdadeiras.
Se qualquer uma das condições for False, o resultado será False.

### **Sintaxe**

```python
condição1 and condição2
```

### **Exemplo**

```python
idade = 20
tem_identidade = True

if idade >= 18 and tem_identidade:
    print("Você pode entrar.")
else:
    print("Entrada negada.")
```

2. or (OU lógico): O operador or retorna True se pelo menos uma das condições for verdadeira.
* O resultado será False apenas se todas as condições forem falsas.

### **Sintaxe**

```python
condição1 or condição2
```

### **Exemplo**

```python
idade = 17
tem_autorizacao = True

if idade >= 18 or tem_autorizacao:
    print("Você pode entrar.")
else:
    print("Entrada negada.")
```

3. not (NÃO lógico): O operador not inverte o valor booleano da expressão.
* Se a expressão for True, o not a transforma em False, e vice-versa.

### **Sintaxe**

```python
not condição
```

### **Exemplo**

```python
tem_ingresso = False

if not tem_ingresso:
    print("Você precisa comprar um ingresso.")
else:
    print("Bem-vindo ao evento.")
```

###Combinações de operadores lógicos

* Os operadores lógicos podem ser combinados para criar expressões mais complexas.
* Quando combinados, a precedência entre eles é importante, com o not sendo avaliado primeiro, seguido de and e, por último, or.

### **Exemplo**

```python
idade = 20
tem_ingresso = True
lista_vip = False

if tem_ingresso and (idade >= 18 or lista_vip):
    print("Você pode entrar.")
else:
    print("Entrada negada.")
```

# Estrutura de Repetição em Python

* As estruturas de repetição, ou laços, são ferramentas fundamentais em Python que permitem executar um bloco de código várias vezes, seja com base em uma condição específica ou um número determinado de iterações.
* Elas são muito úteis quando queremos automatizar tarefas repetitivas, como percorrer elementos de uma lista, executar cálculos ou processar dados.
* Em Python, as principais estruturas de repetição são os laços while e for.

### **Estrutura de repetição while**

* A estrutura while executa um bloco de código repetidamente enquanto uma condição for verdadeira.
* Assim que a condição se tornar falsa, o loop é interrompido.

```python
while condição:
    # Bloco de código a ser repetido
```

### **Exemplo**

```python
contador = 1
while contador <= 5:
    print(f"Contagem: {contador}")
    contador += 1
```

### **Estrutura de repetição for**

* O laço for em Python é amplamente utilizado para iterar sobre uma sequência (como listas, tuplas, strings ou intervalos de números) ou qualquer objeto iterável.
* Ele é geralmente usado quando sabemos de antemão quantas vezes o loop deve ser executado.

### **Sintaxe**

```python
for item in iterável:
    # Bloco de código a ser repetido
```

### **Exemplo**

```python
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(f"Número: {numero}")
```

### Uso comum: iteração sobre sequências

O for é ideal para situações em que você precisa percorrer coleções de dados (listas, dicionários, strings, etc.) ou para um número conhecido de iterações, como:

```python
for letra in "Python":
    print(letra)
```

### Função range() em laços for

* O range() é frequentemente usado com for quando precisamos de uma sequência numérica.
* Ele gera uma sequência de números, que pode ser usada como um iterador.

### **Sintaxe**

```python
for i in range(início, fim, passo):
    # Bloco de código
```

### **Exemplo**

```python
for i in range(1, 6):  # Gera números de 1 a 5
    print(i)
```

### **Estruturas aninhadas**

* Você pode aninhar loops dentro de outros loops para criar repetições mais complexas, como em matrizes ou ao trabalhar com estruturas multidimensionais.

### **Exemplo**

```python
for i in range(1, 4):  # Loop externo
    for j in range(1, 4):  # Loop interno
        print(f"i = {i}, j = {j}")
```

### **Controle do loop: break e continue**

* break: Interrompe o loop imediatamente, independentemente de sua condição.

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

* continue: Pula a iteração atual e passa para a próxima, sem sair do loop.

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

### **Estruturas de repetição com else**

* Python tem uma funcionalidade interessante onde você pode usar a cláusula else com loops.
* O bloco else é executado quando o loop termina normalmente (sem um break).

### **Exemplo**

```python
for i in range(1, 4):
    print(i)
else:
    print("Loop concluído.")
```

### Diferenças entre while e for

* while: É mais útil quando não sabemos quantas vezes o loop será executado, mas sabemos a condição de parada.
* for: É ideal quando temos uma sequência ou sabemos o número de iterações com antecedência.

### Atividade 02

### A empresa Star System Ltda está precisando de uma para o seu sistema de login*.

#### Requisitos da atualização do sistema
*** Bloqueio temporário após 03 tentativas

```python
import time

login = "admin"
senha = "1234"

tentativas_maximas = 3
tempo_bloqueio = 30

tentativas = 0

while tentativas < tentativas_maximas:
    txt_usuario = input("Digite seu nome de usuário: ")
    txt_senha = input("Digite sua senha: ")

    if txt_usuario == login:
        if txt_senha == senha:
            print("Login bem-sucedido!\nBem-vindo ao sistema.")
            break
        else:
            print("Senha incorreta. Tente novamente.")
    else:
        print("Nome de usuário não encontrado.")

    tentativas += 1
    print(f"Tentativa {tentativas} de {tentativas_maximas}")

    if tentativas == tentativas_maximas:
        print("Você excedeu o número máximo de tentativas!!!")
        print(f"O sistema será bloqueado por {tempo_bloqueio} segundos.")
        time.sleep(tempo_bloqueio)
        tentativas = 0
```
