## **Estrutura if**

* A estrutura if avalia uma condio, e se essa condio for verdadeira (True), o bloco de cdigo indentado a seguir ser executado.
* Caso contrrio, o Python pula esse bloco e continua a execuo do programa.

```python
if condio:
    # Cdigo a ser executado se a condio for verdadeira
```

### **Exemplo 01**

```python
idade = 18
if idade >= 18:
    print("Voc  maior de idade.")
```

### **Exemplo 02**

```python
salrio = float(input("Digite o salrio para clculo do imposto: "))

base = salrio
imposto = 0

if base > 3000:
    imposto = imposto + ((base - 3000) * 0.35)
    base = 3000

if base > 1000:
    imposto = imposto + ((base - 1000) * 0.20)

print("Salrio: R$%6.2f\nImposto a pagar: R$%6.2f" % (salrio, imposto))
```

## ***Estrutura if-else***
* A estrutura if-else expande o conceito do if. Se a condio for verdadeira, um bloco  executado.
* Se for falsa, o bloco else ser executado.

### **Sintaxe**

```python
if condio:
    # Cdigo a ser executado se a condio for verdadeira
else:
    # Cdigo a ser executado se a condio for falsa
```

### **Exemplo**

```python
idade = 16
if idade >= 18:
    print("Voc  maior de idade.")
else:
    print("Voc  menor de idade.")
```

### **Exemplo 02**

```python
minutos = int(input("Quantos minutos voc utilizou este ms: "))

if minutos < 200:
    preo = 0.20
else:
    if minutos < 400:
        preo = 0.18
    else:
        preo = 0.15

print("Voc vai pagar este ms: R$%6.2f" % (minutos * preo))
```

## **Estrutura if-elif-else**

* Se houver mltiplas condies a serem verificadas, o elif pode ser utilizado para testar condies adicionais.
* O elif  abreviao de "else if".
* Esta estrutura permite que diferentes blocos de cdigo sejam executados com base em vrias condies.

### **Sintaxe**

```python
if condio1:
    # Cdigo a ser executado se condio1 for verdadeira
elif condio2:
    # Cdigo a ser executado se condio2 for verdadeira
else:
    # Cdigo a ser executado se nenhuma condio for verdadeira
```

### **Exemplo 01**

```python
nota = 75
if nota >= 90:
    print("Voc tirou um A.")
elif nota >= 80:
    print("Voc tirou um B.")
elif nota >= 70:
    print("Voc tirou um C.")
else:
    print("Voc precisa estudar mais.")
```

## Operadores lgicos booleanos

So ferramentas essenciais em programao, usadas para combinar ou modificar expresses lgicas (verdadeiro ou falso). Em Python, eles so empregados para avaliar expresses e retornar um valor booleano: True (verdadeiro) ou False (falso). Esses operadores so muito teis em estruturas de deciso, laos e outros controles de fluxo.

### Principais operadores lgicos booleanos em Python:###
1- and (E lgico): O operador and retorna True apenas se ambas as condies forem verdadeiras.
Se qualquer uma das condies for False, o resultado ser False.

### **Sintaxe**

```python
condio1 and condio2
```

### **Exemplo**

```python
idade = 20
tem_identidade = True

if idade >= 18 and tem_identidade:
    print("Voc pode entrar.")
else:
    print("Entrada negada.")
```

2. or (OU lgico): O operador or retorna True se pelo menos uma das condies for verdadeira.
* O resultado ser False apenas se todas as condies forem falsas.

### **Sintaxe**

```python
condio1 or condio2
```

### **Exemplo**

```python
idade = 17
tem_autorizacao = True

if idade >= 18 or tem_autorizacao:
    print("Voc pode entrar.")
else:
    print("Entrada negada.")
```

3. not (NO lgico): O operador not inverte o valor booleano da expresso.
* Se a expresso for True, o not a transforma em False, e vice-versa.

### **Sintaxe**

```python
not condio
```

### **Exemplo**

```python
tem_ingresso = False

if not tem_ingresso:
    print("Voc precisa comprar um ingresso.")
else:
    print("Bem-vindo ao evento.")
```

# Estrutura de Repetio em Python

* As estruturas de repetio, ou laos, so ferramentas fundamentais em Python que permitem executar um bloco de cdigo vrias vezes, seja com base em uma condio especfica ou um nmero determinado de iteraes.
* Elas so muito teis quando queremos automatizar tarefas repetitivas, como percorrer elementos de uma lista, executar clculos ou processar dados.
* Em Python, as principais estruturas de repetio so os laos while e for.

### **Estrutura de repetio while**

* A estrutura while executa um bloco de cdigo repetidamente enquanto uma condio for verdadeira.
* Assim que a condio se tornar falsa, o loop  interrompido.

```python
while condio:
    # Bloco de cdigo a ser repetido
```

### **Exemplo**

```python
contador = 1
while contador <= 5:
    print(f"Contagem: {contador}")
    contador += 1
```

### **Estrutura de repetio for**

* O lao for em Python  amplamente utilizado para iterar sobre uma sequncia (como listas, tuplas, strings ou intervalos de nmeros) ou qualquer objeto itervel.
* Ele  geralmente usado quando sabemos de antemo quantas vezes o loop deve ser executado.

### **Sintaxe**

```python
for item in itervel:
    # Bloco de cdigo a ser repetido
```

### **Exemplo**

```python
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(f"Nmero: {numero}")
```

### Uso comum: iterao sobre sequncias

O for  ideal para situaes em que voc precisa percorrer colees de dados (listas, dicionrios, strings, etc.) ou para um nmero conhecido de iteraes, como:

```python
for letra in "Python":
    print(letra)
```

### Funo range() em laos for

* O range()  frequentemente usado com for quando precisamos de uma sequncia numrica.
* Ele gera uma sequncia de nmeros, que pode ser usada como um iterador.

### **Sintaxe**

```python
for i in range(incio, fim, passo):
    # Bloco de cdigo
```

### **Exemplo**

```python
for i in range(1, 6):  # Gera nmeros de 1 a 5
    print(i)
```

### **Estruturas aninhadas**

* Voc pode aninhar loops dentro de outros loops para criar repeties mais complexas, como em matrizes ou ao trabalhar com estruturas multidimensionais.

### **Exemplo**

```python
for i in range(1, 4):  # Loop externo
    for j in range(1, 4):  # Loop interno
        print(f"i = {i}, j = {j}")
```

### **Controle do loop: break e continue**

* break: Interrompe o loop imediatamente, independentemente de sua condio.

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

* continue: Pula a iterao atual e passa para a prxima, sem sair do loop.

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

### **Estruturas de repetio com else**

* Python tem uma funcionalidade interessante onde voc pode usar a clusula else com loops.
* O bloco else  executado quando o loop termina normalmente (sem um break).

### **Exemplo**

```python
for i in range(1, 4):
    print(i)
else:
    print("Loop concludo.")
```

### Diferenas entre while e for

* while:  mais til quando no sabemos quantas vezes o loop ser executado, mas sabemos a condio de parada.
* for:  ideal quando temos uma sequncia ou sabemos o nmero de iteraes com antecedncia.

### Atividade 02

### A empresa Star System Ltda est precisando de uma para o seu sistema de login*.

#### Requisitos da atualizao do sistema
*** Bloqueio temporrio aps 03 tentativas

```python
import time

login = "admin"
senha = "1234"

tentativas_maximas = 3
tempo_bloqueio = 30

tentativas = 0

while tentativas < tentativas_maximas:
    txt_usuario = input("Digite seu nome de usurio: ")
    txt_senha = input("Digite sua senha: ")

    if txt_usuario == login:
        if txt_senha == senha:
            print("Login bem-sucedido!\nBem-vindo ao sistema.")
            break
        else:
            print("Senha incorreta. Tente novamente.")
    else:
        print("Nome de usurio no encontrado.")

    tentativas += 1
    print(f"Tentativa {tentativas} de {tentativas_maximas}")

    if tentativas == tentativas_maximas:
        print("Voc excedeu o nmero mximo de tentativas!!!")
        print(f"O sistema ser bloqueado por {tempo_bloqueio} segundos.")
        time.sleep(tempo_bloqueio)
        tentativas = 0
```
